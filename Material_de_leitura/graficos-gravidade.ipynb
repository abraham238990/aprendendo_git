```python
import numpy as np
import matplotlib.pyplot as plt

# Dados experimentais
h = np.array([2.0, 1.5, 1.0, 0.5])              # altura (m)
t = np.array([0.64, 0.55, 0.44, 0.32])          # tempo médio (s)
dt = np.array([0.01, 0.00, 0.017, 0.01])        # desvio padrão do tempo (s)

# Variável t²
t2 = t**2

# Propagação do erro: σ(t²) = 2 t σ(t)
dt2 = 2 * t * dt

# Ajuste linear: h = a * t²
coef = np.polyfit(t2, h, 1)
a = coef[0]

# Gravidade experimental
g = 2 * a

# Gráfico
plt.figure()
plt.errorbar(
    t2, h,
    xerr=dt2,
    fmt='o',
    capsize=4,
    label="Dados experimentais"
)

plt.plot(
    t2,
    a * t2,
    label=f"Ajuste linear\n$g = {g:.2f}$ m/s²"
)

plt.xlabel("t² (s²)")
plt.ylabel("Deslocamento (m)")
plt.title("Determinação da aceleração da gravidade")
plt.grid(True)
plt.legend()
plt.show()

print(f"A aceleração da gravidade experimental é g = {g:.2f} m/s²")
```
